#### I) Confirmed Cases of COVID-19 

```python
# Importing  Confirmed Cases Dataset
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df_confirmed = pd.read_csv(url_confirmed, index_col="Country/Region")
df_confirmed.drop(['Lat', 'Long'], axis=1, inplace=True)
df_confirmed.head(3)

##### Data Preprocessing for the dataset I

# Data for Australian states 
Aust_confirmed = df_confirmed.loc["Australia"]
for i in range(len(Aust_confirmed)):  
      Aust_confirmed.index.values[i]= "Australia_"+ Aust_confirmed.iloc[i]['Province/State']
del Aust_confirmed["Province/State"]  
Aust_confirmed.head(3)

# Country grouping on original dataframe        
gr_confirmed = df_confirmed.groupby("Country/Region").sum()
gr_confirmed.head(3)

# Now adding Australian dataframe
# gr_confirmed = pd.concat([gr_confirmed, Aust_confirmed])

# Adding data for China (1-Jan to 20-Jan from a China CDC publication)
lab=[]
for i in range(1,22):
    lab.append("1/" + str(i) + "/20")
    gr_confirmed.insert(loc=i-1,column=lab[i-1], value=0)
gr_confirmed.loc["China"][0:10] = 20
gr_confirmed.loc["China"][10:21] = 310

# Remove Diamond princess
gr_confirmed = gr_confirmed.drop("Diamond Princess")

# Adding the prefix to  all columns of Confirmed Cases columns, which are dates to distingish them from
# dates under Deaths dataset columns which are also same dates.
gr_confirmed = gr_confirmed.add_prefix('Confirmed Cases on ')
# gr_confirmed.to_csv("C:/Users/user/Downloads/COVID-19 Related Info/confirmed.csv")
gr_confirmed.head(3)
```

#### II) Deaths due to COVID-19

#### IV) Testing for COVID-19 

```python
#  Importing Dataset for Testing
# url_test= 'https://github.com/owid/covid-19-data/raw/master/public/data/testing/covid-testing.xlsx'
url_test= 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/testing/covid-testing-all-observations.csv'
test_df = pd.read_csv(url_test)
                      #index_col="Entity")
test_df.tail(2)
# test_df.to_csv("C:/Users/user/Downloads/COVID-19 Related info/test_df.csv")

#### Data Preprocessing for the dataset III

##### Data Preprocessing for the dataset IV

test_df.shape

test_df.columns

# test_df.drop(['Source URL', 'Source label', 'Notes','3-day rolling mean daily change',
#        '3-day rolling mean daily change per thousand'], axis=1, inplace=True)

#### Step 1: Considering only the max value for Cumulative total per thousand for each Entity for each country for every type of 
#### testing category like tests performed, units unclear, samples tested, cases tested, swabs tested, people tested,
#### inconsistent units (COVID Tracking Project), specimens tested (CDC), samples, samples processed

test_df= (test_df.groupby("Entity").max()[["Cumulative total per thousand"]]) * 1000 # data converted to per M
test_df.head(3)
# test_df.to_csv("C:/Users/user/Downloads/COVID-19 Related info/test_df1.csv")

#### Step 2: Working on "Entity" to make it more useful for our analysis

# When we groupby on a column it becomes the index for the dataframe. Here we did group_by on "Entity", so it is the index for
# the resuting dataframe now.

# Since here the name of the country column is "Entity", which is different from the name of te country column "Country/Region"
# in confirmed cases and deaths dataset
# Also Since the country names are concatenated with strings "- tests performed" which we need to separate from name of the country 
# names in that string. 
# So to perform both of these functions we choose to change the index column  "Entity" to a normal column and then peform the 
# required changes on it and then after applying the changes we can reset it to index again.

# Do the following:
# a) Reset the index to make "Entity" a normal column
# b) Change the name of the index column after resetting
# c) Get the country name from the value under Entity column like: "Argentina" from "Argentina - tests performed"
# d) Set the previous "Entity" column with new name ("Country/Region") and required column values (like, "Argentina") as index

# a)Reset the index to make "Entity" a normal column
test_df = test_df.reset_index("Entity")

# b) Changing the name of the column

# Modifying the name of the "Entity" column to "Country/Region" as in confirmed and deaths datasets to join them using same 
# column with same name

test_df= test_df.rename(columns={"Entity":"Country/Region"})
test_df.columns

test_df.head(3)

# c) To get the country name from the value under Entity column like: "Argentina" from "Argentina - tests performed"
test_df["Country/Region"]= test_df["Country/Region"].str.split(" -", n=1, expand=True)
test_df.head(3)

# d) Resetting the same column with different name and required column values as index
test_df = test_df.set_index("Country/Region")
test_df.head(3)

#### Step 3: Take a maximum for the country out of all categories under testing:

# Since Some countries have more than 1 entry for Cumulative total per thousand maximum value because it is different for
# the categories of testing, like: tests performed, units unclear, samples tested, cases tested, swabs tested, people tested,
# inconsistent units (COVID Tracking Project), specimens tested (CDC), samples, samples processed
test_df= (test_df.groupby("Country/Region").max()[["Cumulative total per thousand"]])

# Checking for difference in index values("Country/Region")  if any in all the datasets (confirmed, deaths, recovered, test) so
# that any discrepancies can either be brought to knowledge and sorted. 

# test_df.shape
# gr_death.shape
# gr_confirmed.shape
# gr_recovered.shape

# check1 = pd.merge(gr_death, gr_confirmed, on=['Country/Region'], how='outer', indicator=True)
# check1.head(2)
# check1[check1['_merge'] == 'left_only']
# check1[check1['_merge'] == 'right_only']

# check2 = pd.merge(gr_death, gr_recovered, on=['Country/Region'], how='outer', indicator=True)
# check2.head(2)
# check2[check2['_merge'] == 'left_only']
# check2[check2['_merge'] == 'right_only']

check = pd.merge(gr_death, test_df, on=['Country/Region'], how='outer', indicator=True)
check.head(2)
# check[check['_merge'] == 'left_only']
check[check['_merge'] == 'right_only']

# Doing this check we get to know that in gr_death and gr_confirmed and gr_recovered has 186 rows each and all values are common
# but in test_df there are just 82 rows out of which following rows are just in test_df and some values out of these are different 
# names for the same country presnt in the other 3 dataframes. So we rename those values and keep other which are not present
# in other 3 dataframes just like that, with NAs for their columns for those countries.
test_df = test_df.rename(index={'United States': 'US', "Czech Republic":"Czechia", "South Korea": "Korea, South"})
gr_death = gr_death.rename(index={'Taiwan*': 'Taiwan'})
gr_confirmed = gr_confirmed.rename(index={'Taiwan*': 'Taiwan'})
gr_recovered = gr_recovered.rename(index={'Taiwan*': 'Taiwan'})

# Validating the results
# check = pd.merge(gr_death, test_df, on=['Country/Region'], how='outer', indicator=True)
# # check[check['_merge'] == 'left_only']
# check[check['_merge'] == 'right_only']

test_df.head(3)
```

##### b) Considering recommended columns in risk index dataset (df_risk_filtered)-> 3 columns

```python
## Analytical Framework for INFORM COVID-19 Risk Index

# COVID-19 HAZARD & EXPOSURE(P2P)- WaSH, (Population)X 2
# Covid-19 Vulnerability  - Movements, Behaviour, (Demographic and Co-morbidities)X 2 
# Covid-19 Lack of coping capacity - IHR, Operational readiness index

# So, all other variables in the dataset are not important for our analysis related to COVID-19,
# The related ones are  
# -> WaSH, (Population)X 2 : Represented by "P2P",  under HAZARD & EXPOSURE category of risk calculation
# -> Movements, Behaviour, (Demographic and Co-morbidities)X 2 : Represented by "Covid-19 Vulnerability", under Vulnerability category of risk calculation
# -> IHR, Operational readiness index : Represented by Covid-19 Lack of coping capacity, under "Lack of coping capacity" category of risk calculation

# ! Note: X 2 means that factor is given higher weightage over other factors in that criteria, e.g. population is given higher weightage
# over WaSH to calculate P2P for HAZARD & EXPOURE

# Therefore, we need only 3 columns for our analysis:
# 1. P2P
# 2. Covid-19 Vulnerability
# 3. Covid-19 Lack of coping capacity
```

```python
df_risk_filtered = df_risk[["P2P", "Covid-19 Vulnerability", "Covid-19 Lack of coping capacity"]]
df1 = gr_confirmed
df2 = gr_death
df3 = gr_recovered
df4 = test_df
df5 = df_risk_filtered
data1 = [df1, df2, df3, df4, df5]

for i in range(1,(len(data1))):
    data1[0] = pd.merge(data1[0], data1[i], how="outer", on="Country/Region")

data_filtered=data1[0]
# data_filtered.to_csv("C:/Users/user/Downloads/COVID-19 Related info/Merged_filtered.csv")
data_filtered.head(3)
```
